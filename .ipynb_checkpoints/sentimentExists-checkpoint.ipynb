```python
import pandas as pd
import numpy as np
from tools.parsers import generalinquirer as generalInquirerParser
from tools.parsers import negation as negationParser 
from tools.sentimentanalysis import preparation
from nltk import word_tokenize
```

```python
prepData = preparation.Preparation()
parserInquirer = generalInquirerParser.GeneralInquirer()
parserNegation = negationParser.Negation()
sentimentDictionaries = parserInquirer.readFileCsv(parserInquirer.combinedFileLoc)
sentences = parserInquirer.readFileCsv(prepData.defaultFileNameSentimentSentences)
```

```python
sentences['textUpper'] = sentences['text'].map( lambda cell: cell.upper())
sentences['wordCount'] = sentences['text'].map( lambda cell: len(word_tokenize(cell)))
```

```python
sentimentDictionaries['entryUpper'] = sentimentDictionaries['entry'].map(lambda cell: cell.upper())
sentimentDictionaries = sentimentDictionaries.fillna("0")
```

```python
positivePolarity = sentimentDictionaries[sentimentDictionaries['priorpolarity'] == 'positive'].reset_index()
negativePolarity = sentimentDictionaries[sentimentDictionaries['priorpolarity'] == 'negative'].reset_index()
weakPositive = sentimentDictionaries[(sentimentDictionaries['priorpolarity'] == 'positive')  & (sentimentDictionaries['type'] == 'weaksubj')].reset_index()
strongPositive = sentimentDictionaries[(sentimentDictionaries['priorpolarity'] == 'positive')  & (sentimentDictionaries['type'] == 'strongsubj')].reset_index()
weakNegative = sentimentDictionaries[(sentimentDictionaries['priorpolarity'] == 'negative')  & (sentimentDictionaries['type'] == 'weaksubj')].reset_index()
strongNegative = sentimentDictionaries[(sentimentDictionaries['priorpolarity'] == 'negative')  & (sentimentDictionaries['type'] == 'strongsubj')].reset_index()
negations = parserNegation.readFileCsv(parserNegation.defaultFileNameProcessed)
```

```python
negations
```

```python
def cleanUpWord(w):
    if w[-1:] in "!?;,.":
        return w[:-1]
    else:
        return w
    
for index, sentence in sentences.iterrows():
    words = [ word for word in [cleanUpWord(w) for w in word_tokenize(sentence['textUpper'])] if len(word) > 0]
    negations['found'] = negations.apply(lambda row: 1 if row['phraseUpper'] in words else 0, axis = 1)
    sentences.loc[index, 'negations'] = negations['found'].sum()
```

```python
def CalculateColumn(column, df, val = None, typeWord = None, normalize = True):
    global sentimentDictionaries
    
    compareVal = column
    if val != None:
        compareVal = val
        
    if typeWord == None:
        lookFor = sentimentDictionaries[sentimentDictionaries[column] == compareVal].reset_index()
    else:
        lookFor = sentimentDictionaries[(sentimentDictionaries[column] == compareVal) & (sentimentDictionaries['type'] == typeWord)].reset_index()
    
    columnName = column+'Count'
    if val != None:
        columnName = columnName + val
    if typeWord != None:
        columnName = columnName + typeWord
    for idx, sent in df.iterrows():
        lookFor['found'] =  lookFor.apply(lambda row: 1 if row['entryUpper'] in sent['textUpper'] else 0, axis = 1)
        found = lookFor['found'].sum()
        if normalize:
            found = lookFor['found'].sum() / float(sent['wordCount'])
        df.loc[idx, columnName] = found
```

```python
sentMixed = sentences[sentences['sentiment-type'] == 0]
sentPolarized = sentences[sentences['sentiment-type'] == 1]

pickedMixed = sentMixed[:30].reset_index()
pickedPolarized = sentPolarized[:120].reset_index()

combinedDf = pd.concat([pickedMixed, pickedPolarized])
combinedDf.drop('level_0', axis=1, inplace=True)
combinedDf = combinedDf.reset_index()
```

```python
CalculateColumn('priorpolarity', combinedDf, 'negative')
CalculateColumn('priorpolarity', combinedDf, 'positive')
CalculateColumn('priorpolarity', combinedDf, 'negative', 'strongsubj')
CalculateColumn('priorpolarity', combinedDf, 'positive', 'strongsubj')
CalculateColumn('priorpolarity', combinedDf, 'negative', 'weaksubj')
CalculateColumn('priorpolarity', combinedDf, 'positive', 'weaksubj')

CalculateColumn('active', combinedDf, None, 'weaksubj')
CalculateColumn('passive', combinedDf, None, 'weaksubj')
CalculateColumn('active', combinedDf, None, 'strongsubj')
CalculateColumn('passive', combinedDf, None, 'strongsubj')
```

```python
CalculateColumn('active', combinedDf)
CalculateColumn('passive', combinedDf)
CalculateColumn('hostile', combinedDf)
```

```python
combinedDf['comparePositiveNegative'] = combinedDf.apply(lambda row: 1 if row['priorpolarityCountpositive'] > row['priorpolarityCountnegative'] else 0, axis = 1)
combinedDf['compareActivePasive'] = combinedDf.apply(lambda row: 1 if row['activeCount'] > row['passiveCount'] else 0, axis = 1)
```

```python
def sumPolarAttributes(row):
    result = row['priorpolarityCountpositive'] + row['activeCountweaksubj'] + row['activeCountstrongsubj']
    result = result - row['priorpolarityCountnegative'] - row['passiveCountweaksubj'] - row['passiveCountstrongsubj'] - row['hostileCount']
    return result

def sumArguing(row):
    result = row['priorpolarityCountpositive'] + row['activeCountweaksubj'] + row['activeCountstrongsubj']
    result = result + row['priorpolarityCountnegative'] + row['passiveCountweaksubj'] + row['passiveCountstrongsubj'] + row['hostileCount']
    return result

combinedDf['diffPolarAttributes'] = combinedDf.apply(lambda row: sumPolarAttributes(row), axis = 1)
combinedDf['arguingAttribute'] = combinedDf.apply(lambda row: sumArguing(row), axis = 1)
combinedDf[['sentiment-type', 'diffPolarAttributes', 'comparePositiveNegative', 'arguingAttribute']]
```

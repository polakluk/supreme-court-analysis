```python
onlyWithAnnotationsIndex = onlyWithAnnotations[:]
onlyWithAnnotationsIndex.loc[onlyWithAnnotations.index, 'index'] = onlyWithAnnotations.index[:]
onlyWithAnnotationsIndex
```

```python
def checkForEachSentence(row, column, annots):
    found = annots[(annots['sentRow'] == row['index']) & (np.invert(annots[column].isnull()))]
    return found.shape[0] > 0    
```

```python
resultsPolarity = onlyWithAnnotationsIndex.apply(lambda row: checkForEachSentence(row, 'polarity', annotations), axis = 1)
```

```python
resultsIntensity = onlyWithAnnotationsIndex.apply(lambda row: checkForEachSentence(row, 'intensity', annotations), axis = 1)
```

```python
resultsIntensity[resultsIntensity == False].shape
```

```python
resultsAttitude = onlyWithAnnotationsIndex.apply(lambda row: checkForEachSentence(row, 'attitude-type', annotations), axis = 1)
```

```python
resultsAttitude[resultsAttitude == False]
```

```python
resultsAttitude[resultsAttitude != -10]
```

```python
finalResultAttitudePolarity = (resultsAttitude == True) & (resultsPolarity == True)
finalResultIntensityPolarity = (resultsIntensity == True) & (resultsPolarity == True)
```

```python
finalResultIntensityPolarity[finalResultIntensityPolarity == False].index
```

```python
finalResultAttitudePolarity[finalResultAttitudePolarity == False].index
```

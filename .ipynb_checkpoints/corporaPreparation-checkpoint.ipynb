```python
with open('./corpora/processed/sentiment-sentences.csv', 'r') as f:
    data = pd.read_csv(f)
```

```python
cleanData = data[data['text'] != 'START']
```

```python
cleanData = cleanData[cleanData['text'] != 'Mr.']
```

```python
from tools.parsers import mpqa as mpqaParser
```

```python
parser = mpqaParser.Mpqa()
annotations = parser.readFileCsv(parser.defaultFileNameProcessedAnnots)
```

```python
for index, sentence in cleanData.iterrows():
    currentAnnots = annotations[(annotations['docName'] == cleanData['docName']) & (annotations['dirName'] == cleanData['dirName']) & (annotations['startByte'] >= cleanData['startByte'] ) & (annotations['endByte'] <= cleanData['endByte']) ]
    cleanData['sentAnnots'][index] = currentAnnots.shape[0]
```
